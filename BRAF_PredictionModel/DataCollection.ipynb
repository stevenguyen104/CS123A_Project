# **Labeling Active, Inactive, and Intermediate coumpunds**

```python
df4 = pd.read_csv('BRAF_data_02.csv')
```

```python
threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    threshold.append("inactive")
  elif float(i) <= 1000:
    threshold.append("active")
  else:
    threshold.append("intermediate")
```

```python
coumpound_class = pd.Series(threshold, name='class')
df5 = pd.concat([df4,coumpound_class], axis=1)
df5
```

```python
df5.to_csv('BRAF_data_03.csv', index=False)
```

# **Cleaning Data**

```python
df6 = pd.read_csv('BRAF_data_03.csv')
```

```python
df6_no_smiles = df6.drop(columns='canonical_smiles')
```

```python
smiles = []

for i in df6.canonical_smiles.tolist():
  cpd = str(i).split('.')
  cpd_longest = max(cpd, key = len)
  smiles.append(cpd_longest)

smiles = pd.Series(smiles, name = 'canonical_smiles')
```

```python
df6_clean_smiles = pd.concat([df6_no_smiles,smiles], axis=1)
df6_clean_smiles
```

# **Normalize the standard_value**
Caps the standard_value at 100000000


```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', axis = 1)

    return x
```

```python
df_norm = norm_value(df6_clean_smiles)
df_norm
```

# **Convert IC50 to pIC50**
pIC50 is the negative logarithmic value of IC50.

Makes the IC50 data more uniformly distributed.

```python
def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', axis = 1)

    return x
```

```python
df_final = pIC50(df_norm)
df_final
```

```python
df_final.to_csv('BRAF_data_04.csv', index=False)
```
